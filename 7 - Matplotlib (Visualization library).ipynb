Matplotlib is a plotting library for the python programming language. It's developed upon NumPy library. 

It alsp provides an object-oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter, etc.

```python
import matplotlib.pyplot as plt

# Writing this line eliminates writing additional statememt to display the plot.
%matplotlib inline 

# importing numpy
import numpy as np
```

## Simple examples

```python
x = np.arange(0, 10)

y = np.arange(11, 21)
```

### Scatter plot

```python
plt.scatter(x, y, c='g') # c determines the color

# setting lables
plt.xlabel("X axis")
plt.ylabel("Y axis")

# setting graph title
plt.title("Graph in 2D")

# save the image
plt.savefig("simple_graph.png")
```

### Line Plot - Straight line

```python
y = x ** 2 # doing basic operation

plt.plot(x, y, c='b', linestyle='dashed') # changing color to blue, changing linestyle
plt.xlabel("X asis")
plt.ylabel("Y axis")
plt.title("Line chart")
```

### Creating Subplots

```python
plt.subplot(2, 2, 1) # passing first subplot(rows, cols, index)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 2) # 2nd col 
plt.plot(x, y, 'b')

plt.subplot(2, 2, 3) # 3rd diagram
plt.plot(x, y, 'g')
```

### Compute x and y for points on sine curve

```python
np.pi
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Creating a subplot
plt.subplot(2, 1, 1)
plt.plot(x, y_sin) # prints the sine wave
plt.title("Sine wave")

# Creating a subplot
plt.subplot(2, 1, 2)
plt.plot(x, y_cos) # prints the sine wave
plt.title("Cos wave")
```

### Bar plot

```python
x = [1, 3, 5, 6, 7]
y = [10, 20, 4, 5, 3]

plt.bar(x, y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar plot")
```

### Box Plot

Very Important - used to find the **percentiles**

Lets look at an example using standard deviation.

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# printing rectangular box plot
plt.boxplot(x=data, vert=True, patch_artist=True)
```

##### Understanding boxplot

In the above chart, bottom line is 0% percentile, blue color first line is 25% percentile, next orange is 50% percentile, and blue ending is 75% percentile and last line is considered as 100% percentile.

```python
data
```

##### Making vertical as False

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# printing rectangular box plot
plt.boxplot(x=data, vert=False)
```

### Piechart

```python
# Data to plot

labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)
```

```python
# Plotting pie chart

plt.pie(x = sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True)

# autopct - shows the percentage on the pie
```
