## Loops and Functions

### Loops

Loops are block of statements which are executed repeatedly until the given condition gets satisfied.

In python, we have two type of loops:
1. `for` loop
2. `while` loop

#### For loop 

This loop is used when we know the start pointer. For example, if we need to print numbers from 1 to n, so the loop initialises with 1. 

for loop syntax is much easy and self-explanatory. In python, we have a special operation where we have the `step` parameter. 

Using the step parameter, we can either do the increment or decrement.

***Syntax***

`for iterator in range(start, end, step):
    statements`


```python
# printing numbers from 1 to 10
for i in range(1, 11, 1): # Note: loop runs from start to end-1
    print(i, end=" ")

print("\n")

# printing numbers from 1 to 10 with step 2
for i in range(1, 11, 2):
    print(i, end=" ")
```

#### Iterating list using for loop

While handling list, we can use different syntax where we can ditectly iterate without specifying the range.

`for iterator in list_variable: 
    print(iterator)`

```python
# initializing a list
li = [53, 32, 21, 67, 90]

# looping over list
for i in li:
    print(i)
```

### While loop 

On the contrary of for loop, `while` loop can be used if we are unaware of the start index or not sure where the loop can be initialized. 

Here the while loop runs until the condition gets false, or we can terminate the loop using the `break` statement. 

***Syntax***

`while(condition): statements`

```python
# printing n numbers backwards from n to 1
n = 10

while(n >= 1):
    print(n, end=" ")
    n-=1
```

#### Using `else` block in while loop

`else` block will execute whenever the while loop condition gets failed or an exception occurs.

```python
# Using the same example shown above

# printing n numbers backwards from n to 1
n = 10

while(n >= 1):
    print(n, end=" ")
    n-=1
else: # gets executed when above condition fails.
    print(0) 
```

#### Function to update age column 

As per the above boxplot, we can see for Pclass 1 the average age is near to 37. 

Similarly for Pclass 2 the average age is near to 29 and Pclass 3 its 24.

```python
def impute_age(cols):

    Age, Pclass = cols[0], cols[1]

    if pd.isnull(Age):

        if Pclass == 1: return 37 

        elif Pclass == 2: return 29

        else: return 24

    else: 
        return Age
```

## Positional Arguments and Keyword arguments

**Positional arguments** - we are passing the variable in the function definition

Eg. `def fun_name(age, name):`

**Keyword arguments** - if the argument passed in the function has a value assigned like key-word pair. 

Eg. `def fun_name(age=34, place='Kakinada'):`

```python
def person(name, age=24):

    print("Hi, My name is {} and age is {}".format(name, age))
```

### Another way using `*args` and `*kwargs`

Whenever we have variable number of arguments, we can use the help of `*args` and `**kwargs`. 

Basically if we pass two lists into one function, it will be considered as positional arguments directly, but if we want to seperate each other then we need to use * for positional and ** for keyword arguments.

```python
def hello(*args, **kwargs):

    print(args)
    print(kwargs)
```

```python
hello("Sai", "Kiran", age=24, place="Kakinada")

# by default positional arguments are taken as one single tuple and 
# and the keyword arguments are taken as Dictionary.
```

```python
pos_arg_li = ['Sai', 'Kiran']
keyword_arg_dict = {'age': 24, 'place': 'Kakinada'}
```

```python
hello(pos_arg_li, keyword_arg_dict)

# As we can see it was considered both of them as positional arguments.
# To seperate both of them, we need to use * and **
```

```python
hello(*pos_arg_li, **keyword_arg_dict)
```

# List Comprehension

Provides a concise way to create lists. It consists of brackets containing an expression followed by a for clause, then zero/more for/if clauses. 

The expressions can be anything meaning you can put all kinds of objects in the lists.

## Generic function to create squares of numbers in list

```python
def compute_squares_from_list(li):

    temp_li = []
    for i in li:
        temp_li.append(i*i)
    
    return temp_li
```

```python
li = [1, 2, 3, 4, 5, 6, 7, 8, 9]

compute_squares_from_list(li)
```

## List Comprehension

Single line of code

```python
squares_list = [i*i for i in li]

print(squares_list)
```

Only computing for even numbers from the list. We can simply add a `if` expression to the end of for loop.

```python
squares__even_list = [i*i for i in li if i%2 == 0]

print(squares__even_list)
```

# List Iterables vs Iterators

## Iterable 
For example, list is an iterable where all the memory gets initialized at once. For example, we need to store 1000 elements at once, so all the data gets stored in memory. 

## Iterator
Consider the same example as above, the whole data 1000 elements will not be stored in memory at once. Instead, if we convert the list into iterator we can extract one element by one using the next function, so everytime we call the next function then only the memory gets initialized. 

> ### In short, Iterator is like a dynamic memory allocation.

## List is Iterable.

It means that all the values are locating in some memory location. While executing the loop, we are iterating over the memory location.

```python
# iterating a list

li = list(range(10))

for i in li: 
    print(i, end=" ")
```

# Using List with iter

`iter()` creates a list_iterator, but to access the next element we need to use `next()` function.

In this case, the whole memory wont be initialized at once compared to list.

**This mostly helps when we have more data and to avoid storing all data at once, we can use the iterator i.e., iter() to extract element one by one.**

<b><u>Note:</u></b> While accessing elements one by one using the next function, and once the elements in the list is completed, then it will throw an error. 

In for loop, this exception handling for this error is already in-place.

```python
# creating iter for the above list

list1 = iter(li)

list1 # creates an list_iterator
```

```python
# Using next(iter) to access the next element

next(list1)
```

## OOPS - Classes, Objects, Methods and Variables

We will take an example of real life example - Car 

```python
class Car:
    
    # __init__ is like a initialization constructor in java
    # self means it directs only for this particular object
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def self_driving(self):

        # Here the self.enginetype gets the value from the above 
        # when its initialized by the constructor
        return "This is a {} car".format(self.enginetype)
```

### Logical Operators

**AND** - returns True only if both condition satisfied

**OR** - returns True even if atleast one condition satisfied

**NOT** - negation of the current result

```python
# Both are same values 
print(True and True) # returns True
print(False and False) # returns False as both are not satisifed

# Both are different values - returns False since one condition is not satisfied.
print(True and False)
print(False and True)
```

```python
# Both are same values 
print(True or True) # returns True
print(False or False) # returns False as both are not satisifed

# Both are different values - returns True since atleast one condition is satisfied.
print(True or False)
print(False or True)
```

# Inheritance

We are taking an exampple of **Car** class.

## Creating Car blueprint

### Types of class variables

1. **Public variables** - Will be directly initialized using self.

2. **Protected variables** - Prefixed with single underscore. Eg. `self._windows = windows`

3. **Private variables** - Prefixed with double underscore. Eg. `self.__windows = windows`

```python
# All the class variables are public, below is the car blueprint

class Car():

    # initializing the constructor
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def drive(self): # We are writing self since it refers to this particular object.
        print ("Person is driving the {} car".format(self.enginetype))
        
```

## Instanting object

```python
car1 = Car(4, 5, "Petrol")
```

```python
car1.drive()
```

## Creating an Audi car using Inherticance property

```python
class audi(Car): # passing the earlier created Car class 

    # initializing constructor
    def __init__(self, windows, doors, enginetype, enableAI): 

        # We can use help of super().__init__(variables...) to fetch the values that were initialized before in the parent class
        super().__init__(windows, doors, enginetype)
        self.enableAI = enableAI

    def selfdriving(self):
        print("Audi supports self driving.")
    
```

### Instantiating object of Audi class

```python
audiQ7 = audi(4, 5, "Petrol", True)
```

```python
audiQ7.selfdriving() # This method is initialized in the child/inherited class.
```

```python
audiQ7.doors
```

```python
audiQ7.drive() # This method is initialized in the parent class
```

## Jointplot

```python
# taking two features tip and total_bill

sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

The points where it is dark is where we have the high concentration. 

In the above jointplot, most people have given tips $2 to $4 and most bill is in the range of $10 to $20.

```python
# taking two features tip and total_bill

sns.jointplot(x='tip', y='total_bill', data=df)
```

# Categorical Plots

The below plots can be used to identify the relationship or analyze the categorical features.

- Countplot()
- Barplot()
- Boxplot() also called "Box and whisker" plot
- Violinplot()


## Countplot()

Shows the counts of observation in each categorical value using bars.

```python
sns.countplot(df['sex'])
```

```python
sns.countplot(x = df['sex'])
```

```python
sns.countplot(df['day'])
```

```python
sns.countplot(df['smoker'])
```

## Barplot

In comparison with Countplot(), in the barplot we need to specify both the x and y values whereas in the countplot(), one axis will contain the count values.

```python
sns.barplot(x=df['sex'], y=df['total_bill'])
```

## Boxplot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

**Data is presented in form of percentiles**

```python
sns.boxplot(x = df['sex'], y = df['total_bill'])
```

### Categorize data based on some categories

`hue` helps to categorize data based on the value passed

```python
sns.boxplot(x=df['total_bill'], y=df['day'], hue=df['smoker'])
```

## Violinplot

Helps us to see both the distribution of data in terms of kernel density estimation (KDE) and the box plot.

```python
sns.violinplot(x=df['total_bill'], y=df['day'])
```

```python
sns.violinplot(x=df['total_bill'], y=df['sex'])
```
