## Importing all the essential libraries

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Dividing the dataset into independent and dependent features

```python
# Syntax - df.iloc[records/rows, columns]

X = df.iloc[:, :-1] # extracting the independent features

y = df.iloc[:, -1] # extracting the dependent feature
```

## Importing Linear Regression Library from sklearn

```python
from sklearn.linear_model import LinearRegression # importing linear regression 

# performs lot of combinations of train and test split so that model be will be accuracte
from sklearn.model_selection import cross_val_score 
```

```python
lin_reg = LinearRegression() # Creating an object
```

**If the mse is 0, then that means the model is accurate**

```python
# Cross validation performs multiple iterations of train and test splits
# cv = 5 means performs 5 times iterations
mse = cross_val_score(lin_reg, X, y, scoring='neg_mean_squared_error', cv=5)

mean_mse = np.mean(mse)

print(mean_mse)
```

## Training the model 

```python
lin_reg.fit(X, y)
```

## Predicting with a new data

```python
lin_reg.predict([[0.00632, 18.0, 2.31, 0, 0.538, 6.575, 65.2, 4.0900, 1, 296.0, 15.3, 396.90, 4.98, 24.0]])
```

## Calculating $\R^2$

```python
lin_reg.score(X, y)
```

# Ridge Regression Implementation

```python
from sklearn.linear_model import Ridge

from sklearn.model_selection import GridSearchCV # used for hyperparameter tuning
```

Creating an object of Ridge library

```python
ridge = Ridge()
```

1e- 15 means 0.0000000000000001 

1e-10 means 0.000...01 (Ten number of 1's after 1)

```python
params = {'alpha': [1e-15,1e-10,1e-8, 1e-3, 1e-2, 1, 5, 10, 20]}
```

```python
ridge_regressor = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=5)
```

Training the model

```python
ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_) # prints the best alpha

# best score shows the performance of the model
print(ridge_regressor.best_score_) # here in the linear regression we got -1 now we got -2 so Ridge regression is not suitable.
```

# Lasso Regression Implementation

```python
from sklearn.linear_model import Lasso

from sklearn.model_selection import GridSearchCV # used for hyperparameter tuning
```

```python
lasso = Lasso()
```

```python
params = {'alpha': [1e-15,1e-10,1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 40, 50, 60, 70]}
```

```python
lasso_regressor = GridSearchCV(lasso, params, scoring='neg_mean_squared_error', cv=5)
```

```python
lasso_regressor.fit(X, y)
```

```python
print(lasso_regressor.best_params_) # prints the best alpha

print(lasso_regressor.best_score_) 
```

# Logistic Regression Implementation

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

Independent features

```python
X = pd.DataFrame(df['data'], columns=df['feature_names'])
```

```python
X.head()
```

Dependent features

```python
y = pd.DataFrame(df['target'], columns=['Target'])
```

```python
y
```

```python
y['Target'].value_counts() 
```

Train Test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44)
```

```python
params = [{'C' : [1,5,10]}, {'max_iter': [100, 150]} ]
```

```python
model = LogisticRegression(C = 100, max_iter=100)
```

```python
GridSearchCV(model, param_grid=params, scoring='f1', cv=5)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

y_pred
```

Importing Confusion matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
