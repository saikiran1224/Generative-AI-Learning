## Loading a sample dataset

```python
# Loading dataset

df = sns.load_dataset("tips") # dataset from restaurant where customer paid tips.
```

# Problem Statement 

Create a model where we need to find the **tip** based on the other features. 

Here, **tip** is the dependent feature, and all the remaining features **(total_bill, sex, smoker, day, time and size)** are considered to be independent. 

# Numerical plots: (Works only for numerical values)

**Correlation heatmap** - Helps to identify the correlation among the features

**Jointplot** - Useful to visualize dataset having 2 independent features.

**Pairplot** - Useful to visualize dataset having >3 independent features.

**Distplot** - similar to the histogram we used in matplotlib. Shows the density of the values.

## Correlation Heatmap 

This uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix. 

Helps us to find the inter-relation between features. This only works for numerical features (integer and float) and not works for categorical features.

```python
df.dtypes
```

Correlation heatmap only works for float and int datatypes.

```python
# excluding integer datatypes

df_2 = df.select_dtypes(exclude=['category'])

df_2.dtypes
```

```python
# Matrix of correlation

df_2.corr()
```

### Understanding the above correlation

- If the total_bill increases, both tip and size is also increasing positively. 

```python
# displaying the heatmap by passing the matrix of correlation

sns.heatmap(df_2.corr())
```

If in the correlation map, if the values are same for two independent variables, then we can choose one variable instead of using both of them.

## Jointplot

Helps us to do Univariate analysis and Bivariate analysis.

```python
# taking two features tip and total_bill

sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

The points where it is dark is where we have the high concentration. 

In the above jointplot, most people have given tips $2 to $4 and most bill is in the range of $10 to $20.

```python
# taking two features tip and total_bill

sns.jointplot(x='tip', y='total_bill', data=df)
```

## Pairplot 

Used when we want to analyze more than two independent features. Pairplot does the permutation and combination of all the features and displays all the plots.

```python
sns.pairplot(df)
```

Specifying the feature based on which plots needs to be calculated

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='smoker')
```

## Displot 

Helps us to check the distribution of the columns feature. 

```python
sns.displot(df['tip'])
```

### KDE means Kernel Density Estimation. 

When we use KDE, it displays the value in form of percentage instead of count.


```python
sns.displot(df['tip'], kind='kde') 
```
