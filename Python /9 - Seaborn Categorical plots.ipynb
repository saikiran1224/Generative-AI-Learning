# Seaborn Categorical Plots 

In the previous notebook, the plots which are used - Jointplot (univariate and bivariate analysis), Pairplot (trivariate analysis) and jointplot (histogram) can only be used for Numerical values (integer and float values).

In this notebook, we will discuss the plots which can be used for analyzing the categorical features (object datatype).

## Importing seaborn and pandas

```python
import seaborn as sns
import pandas as pd
```

## Loading restaurant tips data

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

# Categorical Plots

The below plots can be used to identify the relationship or analyze the categorical features.

- Countplot()
- Barplot()
- Boxplot() also called "Box and whisker" plot
- Violinplot()


## Countplot()

Shows the counts of observation in each categorical value using bars.

```python
sns.countplot(df['sex'])
```

```python
sns.countplot(x = df['sex'])
```

```python
sns.countplot(df['day'])
```

```python
sns.countplot(df['smoker'])
```

## Barplot

In comparison with Countplot(), in the barplot we need to specify both the x and y values whereas in the countplot(), one axis will contain the count values.

```python
sns.barplot(x=df['sex'], y=df['total_bill'])
```

## Boxplot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

**Data is presented in form of percentiles**

```python
sns.boxplot(x = df['sex'], y = df['total_bill'])
```

### Categorize data based on some categories

`hue` helps to categorize data based on the value passed

```python
sns.boxplot(x=df['total_bill'], y=df['day'], hue=df['smoker'])
```

## Violinplot

Helps us to see both the distribution of data in terms of kernel density estimation (KDE) and the box plot.

```python
sns.violinplot(x=df['total_bill'], y=df['day'])
```

```python
sns.violinplot(x=df['total_bill'], y=df['sex'])
```
