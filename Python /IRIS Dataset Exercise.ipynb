# IRIS Dataset - Analysis using Seaborn

## Importing seaborn library

```python
import seaborn as sns
```

## Loading IRIS dataset 

```python
df = sns.load_dataset('iris')
```

## Displaying `df.head()`

```python
df.head()
```

## Observations of dataset

`species` is the dependent variable.

`sepal_length`, `sepal_width`, `petal_length` and `petal_width` are the independent variables.

Since all the above dataset features are Numerical values, we will analyze using the various numerical plots. Below are the plots that we will use:
- Correlation Heatmap()
- Jointplot (for univariate and bivariate analysis)
- Pairplot (>2 feature analysis)
- Distplot (Histogram)

## Correlation Heatmap

```python
df.dtypes
```

### Excluding categorical features from the df 

```python
df_2 = df.select_dtypes(exclude=['object'])
```

```python
df_2.dtypes
```

```python
df_2.corr()
```

```python
sns.heatmap(df_2.corr())
```

## Jointplot

```python
sns.jointplot(x = df['species'], y = df['petal_width'])
```

```python
sns.jointplot(x = df['petal_width'], y = df['petal_length'],kind='hex')
```

## Pairplot 

Performs permutations and combinations and displays the charts

```python
sns.pairplot(df)
```

## Distplot

Displays historgram-like charts

```python
sns.displot(df['species'])
```

```python
sns.displot(df['petal_length'])
```

```python
sns.displot(df['petal_length'], kind='kde') # displays in percentage
```
